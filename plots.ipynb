```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('students_in_class.csv')
```

```python
data
```

```python
mStud = max(data['Enrolment in Class I to V - Boys'][:-1].max(), data['Enrolment in Class I to V - Girls'][:-1].max())
plt.figure(figsize=(10, 7))
plt.scatter(data['Enrolment in Class I to V - Boys'][:-1], data['Enrolment in Class I to V - Girls'][:-1], color='red')
plt.plot([0,mStud],[0,mStud], color='blue', linestyle='dotted')
plt.xlabel('Number of boys enrolled in class I-V')
plt.ylabel('Number of girls enrolled in class I-V')
plt.title('Comparison of boys and girls enrolled for different states')
plt.savefig('Scatter')
```

```python
plt.figure(figsize=(10, 7))
plt.boxplot([data['Enrolment in Class I to V - Boys'][:-1], data['Enrolment in Class I to V - Girls'][:-1]], labels=['Boys','Girls'])
plt.ylabel('Students enrolled')
plt.title('Boys and Girls enrolled for different states')
plt.savefig('Boxplot boys girls')
```

```python
plt.figure(figsize=(10, 7))
plt.boxplot([data['Enrolment in Class I - Total'][:-1], data['Enrolment in Class II - Total'][:-1],
            data['Enrolment in Class III - Total'][:-1], data['Enrolment in Class IV - Total'][:-1],
            data['Enrolment in Class V - Total'][:-1]], labels=['Class I','Class II','Class III','Class IV','Class V'])
plt.ylabel('Students enrolled')
plt.title('Students enrolled for different states for different classes')
plt.savefig('Boxplot classes')
```

```python
plt.figure(figsize=(10, 7))
labels = ['Class I','Class II','Class III','Class IV','Class V']
X_axis = np.arange(len(labels))

plt.bar(X_axis - 0.2,
        [data['Enrolment in Class I - Boys'][:-1].sum(), data['Enrolment in Class II - Boys'][:-1].sum(),
            data['Enrolment in Class III - Boys'][:-1].sum(), data['Enrolment in Class IV - Boys'][:-1].sum(),
            data['Enrolment in Class V - Boys'][:-1].sum()],
       0.4, label='Boys')
plt.bar(X_axis + 0.2,
        [data['Enrolment in Class I - Girls'][:-1].sum(), data['Enrolment in Class II - Girls'][:-1].sum(),
            data['Enrolment in Class III - Girls'][:-1].sum(), data['Enrolment in Class IV - Girls'][:-1].sum(),
            data['Enrolment in Class V - Girls'][:-1].sum()],
       0.4, label='Girls')
plt.xticks(X_axis, labels)
plt.xlabel("Classes")
plt.ylabel('Students enrolled')
plt.title('Boys and Girls enrolled for different classes')
plt.legend()
plt.savefig('barplot')
```
